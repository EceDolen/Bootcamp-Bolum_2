```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
from scipy import stats
import seaborn as sns
```

dağılım = np.random.dağılım_adı([],[]) kodunu kullanarak, dağılım_adı yazan yere seçtiğiniz bir dağılımı ve parantez içinde [] şeklinde gösterilen boşlukları yine kendi seçiminiz doğrultusunda doldurun.

```python
#mu, sigma, n = 0.5, 0.1, 10000
dagilim = np.random.normal(0.5,0.1,1000)
```

a) Verinin histogramını çiziniz

```python
plt.figure(figsize=(3,2), dpi = 150)
count, bins, ignore = plt.hist(dagilim, 50, density=True)
plt.plot(bins, 1/(0.1 * np.sqrt(2 * np.pi))*np.exp( - (bins - 0.5)**2 / (2 * 0.1**2) ),
         linewidth=3, color='y')
plt.show()
```

b) Q-Q grafiğini çizip yorumlayınız

```python
dagilim.sort()


plt.figure(figsize=(5,4), dpi = 100)
nsample = 100
res = stats.probplot(dagilim, plot=plt)
plt.title("Q-Q Çizgesi")
plt.xlabel("Teorik Değerler")
plt.ylabel("Randon Dagilim Veriler")
plt.show()
```

```python
from scipy.stats import norm
sns.distplot(dagilim, fit = norm)

#siyah cizgi bu verinin varyans ve ortalamasına göre ideal olan
```
